```python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5" #"claude-sonnet-5"
```

```python
# define helper functions
def add_user_message(messages, content):
    messages.append({"role": "user", "content": content})

def add_assistant_message(messages, content):
    messages.append({"role": "assistant", "content": content})

def chat(messages):
    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return response.content[0].text
```

```python
# Make an empty list of messages
messages = []

# Add in the initial user question
add_user_message(messages, "Why are openings that start with 1. e4 e5 known as \"open games\" in chess?  Answer briefly, at most two sentences.")

# Pass the list of messages into 'chat' to get an answer
answer = chat(messages)

# Add the assistant's answer to the list of messages
add_assistant_message(messages, answer)

# Add a follow-up user question
add_user_message(messages, "What are some common openings that start with 1. e4 e5? Answer in a single sentence.")

# Call 'chat' again to get an answer to the follow-up question
answer = chat(messages)
answer
```

```python
# Creating a simple chat loop

# Start with empty list of messages
messages = []

# Infinite loop to run bot forever
while True:
    # Get user input
    user_input = input("> ")
    print ("> ", user_input)

    # Add user input to list of messages
    add_user_message(messages, user_input)

    # Call Claude API to generate answer
    answer = chat(messages)

    # Print answer
    print("---")
    print(answer)
    print("---")

    # Add answer to list of messages
    add_assistant_message(messages, answer)

```
