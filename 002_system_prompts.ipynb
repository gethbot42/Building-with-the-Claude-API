```python
# Load env variables and create API Client
from dotenv import load_dotenv

load_dotenv()

# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5" #"claude-sonnet-5"

```

```python
# define helper functions
def add_user_message(messages, content):
    messages.append({"role": "user", "content": content})

def add_assistant_message(messages, content):
    messages.append({"role": "assistant", "content": content})

# Modified chat function from previous notebook
# Adding system propmt to control Claude output behavior
def chat(messages):
    system = """
    You are a patient math tutor.
    Do not directly answer a student's questions.
    Guide them to a solution step by step.
    """

    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
        system=system
    )
    return response.content[0].text
```

```python
# Modifying chat function to take system argument
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if system:
        params["system"] = system
    
    message = client.messages.create(**params)

    return message.content[0].text
```

```python
# Now we can add an optional system prompt to our chat function dynamically

# System prommpt exercise: have claude produce a python function that checks a string for dup characters
# Claude will typically add extra explanations and comments... we need to write a system prompt to prevent this

messages = []
add_user_message(messages, "Write a python function that checks a string for duplicate characters.")

# The key in this system prompt is assigning Claude a role, then telling it how the response should be structured.
system = """
You are a software engineer.
When asked to write code or functions, only return the code, no other text.
Only add comments if parts of the code are unclear or there is special logic used.
"""

answer = chat(messages, system)

answer
```

```python
# Adding temperature to the chat function
def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system
    
    message = client.messages.create(**params)

    return message.content[0].text
```

```python
messages = []

add_user_message(
    messages,
    "Generate a one sentence movie idea"
)

# lower temps --> more consistant output
# higher temps  --> more suited for creative work
answer = chat(messages, temperature=0.0)

answer
```
